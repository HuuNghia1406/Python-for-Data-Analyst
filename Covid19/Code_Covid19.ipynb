 ## I. DATA PROCESSING

 IMPORT CÁC THƯ VIỆN CẦN THIẾT

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

 ĐỌC TẬP DỮ LIỆU BẰNG THƯ VIỆN PANDAS

```python
covid19 = pd.read_csv('D:/Python-for-Data-Analyst/DataSet/covid19_Confirmed_dataset.csv')
covid19.sample(10)
```

XEM SỐ LƯỢNG HÀNG, CỘT

```python
covid19.shape
```

XOÁ NHỮNG CỘT KHÔNG CẦN THIẾT

```python
covid19.drop(['Long','Lat','Province/State'],axis='columns',inplace=True)
covid19.sample(3)


```

- Chúng ta đã xoá đi 2 cột là Long và Lat

GOM NHÓM THEO CỘT Country/Region

```python
covid19_groupby_CountryRegion = covid19.groupby('Country/Region').sum()
covid19_groupby_CountryRegion.head()
```

XEM LẠI SỐ CỘT, DÒNG

```python
covid19_groupby_CountryRegion.shape
```

BIỂU ĐỒ THỂ HIỆN SỐ MẮC COVID QUA TỪNG THỜI ĐIỂM , LẤY VÍ DỤ 2 NƯỚC LÀ CHINA VÀ ITALIA

```python
covid19_groupby_CountryRegion.loc['China'].plot()
covid19_groupby_CountryRegion.loc['Italy'].plot()
```

GỈA SỬ CHÚNG TA CHỈ MÚN HIỂN THỊ 3 NGÀY ĐẦU MẮC COVID CỦA CHINA 

```python
covid19_groupby_CountryRegion.loc['China'][:3].plot()
```

```python
covid19_groupby_CountryRegion.loc['China'].diff().plot()
```

- Ở đây phương thức diff() cho biết sự biến thiên của giá trị trước và giá trị liền kề sau nó là bao nhiêu

NẾU TA MÚN XEM SỰ BIẾN THIÊN CAO NHẤT LÀ BAO NHIÊU, GIẢ SỬ 2 NƯỚC LÀ ITALY VÀ INDIA

```python
covid19_groupby_CountryRegion.loc['India'].diff().max()
```

```python
covid19_groupby_CountryRegion.loc['Italy'].diff().max()
```

TẠO CỘT MỚI CHỨA SỰ BIẾN THIÊN CAO NHẤT THEO TỪNG THỜI GIAN CỦA TỪNG QUỐC GIA

```python
countries  = list(covid19_groupby_CountryRegion.index)
max_infection_rates = []
# for c in countries:
#     max_infection_rates.append(covid19_groupby_CountryRegion.loc[c].diff().max())
for index, row in covid19_groupby_CountryRegion.iterrows():
    max_infection_rates.append(row.diff().max())
```

- Đầu tiên lại ra 1 list để chứa danh sách các quốc gia
- Sau đó duyệt qua từng hàng để tìm sự biến thiên cao nhất
- Cuối cùng kết quả được lưu ngược lại mảng rỗng đã tạo từ trước

```python
covid19_groupby_CountryRegion['max_infection_rates'] = max_infection_rates
covid19_groupby_CountryRegion.head()
```

- Tạo ra cột mới chứa giá trị của mảng đã chứa kết quả trước đó

BÂY GIỜ TÔI MÚN TẠO RA 1 DATAFRAME VỚI CHỈ DUY NHẤT CỘT VỪA TẠO

```python
covid19_max_infection_rates = pd.DataFrame(covid19_groupby_CountryRegion['max_infection_rates'])
covid19_max_infection_rates.sample(10)
																			
```
